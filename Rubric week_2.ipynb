```python
def bracelet(a, b):
    """Return the repeating cycle starting from (a,b)."""
    seen = {}
    seq = []

    while (a, b) not in seen:
        seen[(a, b)] = len(seq)
        seq.append(a)
        a, b = b, (a + b) % 10

    start = seen[(a, b)]
    return tuple(seq[start:])  # the cycle


# Collect bracelets
bracelets = {}

for a in range(10):
    for b in range(10):
        cycle = bracelet(a, b)
        bracelets.setdefault(cycle, []).append((a, b))


# Display results
print(f"Number of distinct bracelets: {len(bracelets)}\n")

for cycle, pairs in sorted(bracelets.items(), key=lambda x: len(x[0])):
    print(f"Cycle length {len(cycle)}:")
    print("  cycle:", cycle)
    print("  starting pairs:", pairs)
    print()
```
